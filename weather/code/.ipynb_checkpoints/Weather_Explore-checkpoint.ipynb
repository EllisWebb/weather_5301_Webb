In order to confidently do a t-test we must ensure the data is normally distrubuted using a histogram. As we can see it is clearly not as our data is skewed left. That being said, due to the CLT and the nature of the size of our data sets, we can still use the t-test as a reliable indicator. 

```python
plt.figure(figsize=(20, 5))

sns.histplot(data=df.loc[df['month'] == 1], x='precipitation', hue='city')

plt.xlabel('Precipitation (inches)', fontsize=18)
plt.ylabel('# of Days', fontsize=18)
plt.title('Precipitation in January', fontsize=20)

plt.tick_params(labelsize=15)

plt.show()
```

Importing scipy libraries in order to conduct a t-test. Since I was, at first, unable to weed out a good portion of my NaN values during cleaning we can see that many of our t-tests didnt proceed correctly. However, this issue has now been resolved.

```python
from scipy import stats
```

```python
significance_level = 0.05
significantly_different = np.zeros(12)

for month in range(1, 13):
    sea_data = df.loc[(df['city'] == 'SEA') & (df['month'] == month), 'precipitation'] 
    van_data = df.loc[(df['city'] == 'VAN') & (df['month'] == month), 'precipitation'] 

    t_statistic, p_value = stats.ttest_ind(sea_data, van_data, equal_var=False)

    if p_value < significance_level:
        significantly_different[month-1] = 1

    print(f"Month {month}:")
    print(f" t_statistic = {t_statistic:.2f}:")
    print(f" p_value t test = {p_value:.3f}:")
    print("-" * 20)
```

By adding markers to our graph we can visually see significant findings within our data set. 

```python
plt.figure(figsize=(20,4))

sns.barplot(data=df, x='month', y='precipitation', hue = 'city', alpha = 0.75)

plt.xlabel('Month', fontsize = 15)
plt.ylabel('Precipitation (Inches)', fontsize=15)
plt.title('Mean Precipitation By Month', fontsize=20)

plt.tick_params(labelsize=15)

plt.xticks(ticks=range(12), labels=month_names)

for month in range(12):
    if significantly_different[month] == 1:

        plt.text(month, 0.3, '!', ha='center', fontsize=30)

plt.show()
```
